```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
q1 = {'€1.50':1,'€2':5,'€2.50':1,'€3':3,'€4':2}
plt.bar(q1.keys(),height=q1.values())
plt.xlabel('Price')
plt.ylabel('Number of responses')
plt.title('Price respondents would pay for product')
plt.show()
```

Mean Pay Price: €2.58

```python
marketing = {'Instagram':5,'Social Media':6,'Tik Tok':2,'TV':3,'Supermarket Catalogue':1}
marketing_methods = list(marketing.keys())
counts = list(marketing.values())
plt.pie(counts,labels = marketing_methods, autopct='%1.1f%%')
plt.title('Most Effective Marketing Channel')
plt.show()
```

```python
flavours = {'Chicken':2,'French Onion':1,'Mushroom':1,'Tomato':3,'Carrot and Corriander':2,
            'Potato and Leek':1,'Vegetable':3}
plt.bar(flavours.keys(),height = flavours.values())
plt.xticks(rotation = 80)
plt.yticks([i for i in range(0,4,1)])
plt.xlabel('Flavour')
plt.ylabel('Responses')
plt.title('Survey Respondants Favourite Flavour')
plt.show()
```

```python
packaging = {'0-3':0,'4-7':3,'8-10':10}
plt.bar(packaging.keys(),height = packaging.values())
plt.xlabel('Score Range')
plt.ylabel('Number of Responses')
plt.title('Responses of How Well Packaging Reflects Soup')
plt.show()
```
